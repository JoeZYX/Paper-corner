**Statlog Shuttle Radiator data 穿梭散热器数据**

To classify potential anomalies in the radiator position onboard Space Shuttles.

There are 7 types of exceptions, which are coded as follows : 
- Rad Flow 1 (正常状态）
- Fpv Close 2
- Fpv Open 3
- High 4
- Bypass 5
- Bpv Close 6
- Bpv Open 7

The shuttle dataset contains 9 attributes all of which are numerical. But the meaning of specific features is not stated

```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mlp

import seaborn as sns
```

# Load Raw Data

```python
# Column names to be used for dataframe
train_col_names = ['A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'A7', 'A8', 'A9', 'Class']
test_col_names = ['A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'A7', 'A8', 'A9']

training_data = pd.read_csv("data/train.csv", index_col=[0],header=[0])
testing_data = pd.read_csv("data/test.csv", header=[0],index_col=[0])

training_data.columns = train_col_names
testing_data.columns = test_col_names

print("The shape of training dataset is ", training_data.shape)
print("The shape of test dataset is ", testing_data.shape)
```

# dummy models

```python
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split


from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, f1_score, recall_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import plot_confusion_matrix
```

```python
X = training_data.drop('Class', axis = 1)
X = X.values.astype("float")

y = training_data['Class']

rb_scaler = RobustScaler()
X_transformed = rb_scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_transformed, y, test_size = 0.30, stratify = y)
```

```python

rf_clf = RandomForestClassifier(n_estimators=500)


rf_clf.fit(X_train, y_train)


y_pred_rf = rf_clf.predict(X_test)


# Get model metrics to gauge model performance-
accuracy_rf = accuracy_score(y_test, y_pred_rf)
precision_rf = precision_score(y_test, y_pred_rf, average='macro')
recall_rf = recall_score(y_test, y_pred_rf, average='macro')

print("\nRandomForestClassifier (n_estimators = 500) model metrics are:")
print("Accuracy = {0:.4f}, Precision = {1:.4f} & Recall = {2:.4f}\n".format(accuracy_rf, precision_rf, recall_rf))
```

```python
print(classification_report(y_test, y_pred_rf))
```

```python

# Get feature importance using 'RandomForestClassifier'-
fi_rfc = pd.DataFrame({'feature_importance_scores': rf_clf.feature_importances_, 'attributes': train_col_names[:-1]})

# Sort in descending order-
fi_rfc.sort_values('feature_importance_scores', ascending=False, inplace=True)

fi_rfc
```

```python
importances = rf_clf.feature_importances_
std = np.std([tree.feature_importances_ for tree in rf_clf.estimators_],
             axis=0)
indices = np.argsort(importances)

# Plot the feature importances of the forest
plt.figure(figsize=(15,5))
plt.title("Feature importances")
plt.barh(range(X.shape[1]), importances[indices], xerr=std[indices], align="center")
yticks = [train_col_names[i] for i in indices]
plt.yticks(range(len(indices)),yticks)
plt.ylim([-1, X.shape[1]])
plt.show()
```

# Anomaly Detection

The original Statlog (Shuttle) dataset from UCI machine learning repository is a multi-class classification dataset with dimensionality 9. Here, the training and test data are combined. The smallest five classes, i.e. 2, 3, 5, 6, 7 are combined to form the outliers class, while class 1 forms the inlier class. Data for class 4 is discarded.

```python
from scipy.io import loadmat
data_anomaly = loadmat("data/shuttle.mat")
```

```python

Anamaly_df = pd.DataFrame(data_anomaly["X"])
Anamaly_df.columns = training_data.columns[:9]
Anamaly_df["Class"] = data_anomaly["y"]
```
